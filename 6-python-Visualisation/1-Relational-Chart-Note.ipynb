# **lineplot**

```python


# Load example dataset
flights = sns.load_dataset('flights')
flights.head()
```

```python
print(flights)
```

```python
sns.lineplot(data=flights, x='year', y='passengers')
plt.title("Basic Line Plot")
plt.show()
```

### 2ï¸âƒ£Â Line Plot withÂ hue

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
plt.title("Line Plot with Hue")
plt.show()
```

### 3ï¸âƒ£Â Line Plot withÂ style

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month', style='month')
plt.title("Line Plot with Hue and Style")
plt.show()
```

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month', size='month', sizes=(1, 5))
plt.title("Line Plot with Variable Line Widths")
plt.show()
```

```python

sns.lineplot(data=flights, x='year', y='passengers', hue='month', palette='tab10')
plt.title("Line Plot with Custom Palette")
plt.show()

```

```python

months_order = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December']

sns.lineplot(
    data=flights,
    x='year',
    y='passengers',
    hue='month',
    hue_order=months_order,
    palette='tab10'
)
plt.title("Line Plot with Ordered Hue")
plt.show()

```

```python

sns.lineplot(
    data=flights,
    x='year',
    y='passengers',
    hue='month',
    style='month',
    markers=True
)
plt.title("Line Plot with Markers")
plt.show()

```

```python

sns.lineplot(
    data=flights,
    x='year',
    y='passengers',
    hue='month',
    style='month',
    dashes=False  # Use solid lines only
)
plt.title("Line Plot with Solid Lines")
plt.show()

```

```python

sns.lineplot(
    data=flights,
    x='year',
    y='passengers',
    hue='month',
    style='month',
    size='passengers',
    sizes=(1, 5),
    markers=True,
    dashes=True
)
plt.title("Comprehensive Line Plot")
plt.show()

```

```python

sns.lineplot(data=flights, x='year', y='passengers', hue='month', alpha=0.7)
plt.title("Line Plot with Transparency")
plt.show()
```

## Rel plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
tips = sns.load_dataset('tips')
fmri = sns.load_dataset('fmri')
penguins = sns.load_dataset('penguins')
```

```python
tips.head()
```

```python
fmri.head()
```

```python
penguins.head()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
plt.show()
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', errorbar='sd')
plt.show()
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', errorbar=None)
plt.show()
```

```python
sns.relplot(data=tips,
            x='total_bill', y='tip',
            hue='sex', style='smoker',
            size='size', 
            kind='line')

plt.show()
```

```python
uk_election = pd.read_excel('uk_election.xlsx', sheet_name=5)
uk_election 
```

```python
def clean_column_names(df):
    df.columns = (df.columns
                  .str.strip()
                  .str.lower()
                  .str.replace(' ', '_')
                  .str.replace('(', '')
                  .str.replace(')', '')
                 )
    return df


uk_election = clean_column_names(uk_election)
uk_election.columns.to_list()
```

```python
uk_election.columns.values[2] = 'year_2019'
uk_election.columns.values[3] = 'year_2024'
uk_election.columns.to_list()
```

```python
ax, fig = plt.subplots(figsize=(10, 6))
sns.relplot(data= uk_election,
            y='year_2019', x='region',
            kind='scatter', ax=ax)
```
