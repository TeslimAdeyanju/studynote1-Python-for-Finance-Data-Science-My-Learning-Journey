```python
# Analyze each column
for col_name, col_data in df.items():
    print(f"\n=== {col_name} ===")
    print(f"Data type: {col_data.dtype}")
    print(f"Non-null count: {col_data.count()}/{len(col_data)}")
    if col_data.dtype == 'object':
        print(f"Unique values: {col_data.nunique()}")
```

```python
quality_report = {}

for col_name, col_data in df.items():
    quality_report[col_name] = {
        'null_count': col_data.isnull().sum(),
        'unique_count': col_data.nunique(),
        'dtype': str(col_data.dtype)
    }
```

```python
print(list(quality_report.keys()))  # Converts the dictionary keys to a list and prints them
```

```python
# Apply different transformations based on dtype
for col_name, col_data in df.items():
    if col_data.dtype == 'object':
        df[col_name] = col_data.str.upper()
    elif col_data.dtype in ['int64', 'float64']:
        df[col_name] = col_data.fillna(col_data.median())
        
print(df.head())
```

```python
for col_name, col_data in df.items():
    if col_data.dtype == 'object':
        print(f"\n=== {col_name} ===")
        print(col_data.value_counts(dropna=False))
```

## Using .apply() for Z-score Calculation

Here are several ways to calculate z-scores using `.apply()`:

```python
# Method 1: Using apply with lambda function (column-wise)
def calculate_zscore(series):
    """Calculate z-score for a series"""
    return (series - series.mean()) / series.std()

# Apply the function to each numeric column
zscore_df = numeric_cols.apply(calculate_zscore).fillna(0)

# Add suffix to column names
zscore_df.columns = [col + '_zscore' for col in zscore_df.columns]

# Concatenate with original DataFrame
df_with_zscores = pd.concat([df, zscore_df], axis=1)

print("Z-scores calculated using apply:")
print(df_with_zscores[['age', 'age_zscore', 'fare', 'fare_zscore']].head())
```

```python
# Method 2: Using apply with lambda (more concise)
zscore_lambda = numeric_cols.apply(lambda x: ((x - x.mean()) / x.std()).fillna(0))
zscore_lambda.columns = [col + '_zscore_v2' for col in zscore_lambda.columns]

print("Z-scores using lambda apply:")
print(zscore_lambda.head())
```

```python
# Method 3: Direct application to existing DataFrame (modifying in place)
# This replicates your original approach but using apply

for col in numeric_cols.columns:
    df[col + '_zscore_v3'] = numeric_cols[col].apply(
        lambda x: (x - numeric_cols[col].mean()) / numeric_cols[col].std()
    ).fillna(0)

print("Z-scores added to original DataFrame:")
print(df[['age', 'age_zscore_v3', 'fare', 'fare_zscore_v3']].head())
```

```python
# Method 4: Using scipy.stats for standardization (bonus method)
from scipy import stats

# Apply scipy's zscore function to each column
zscore_scipy = numeric_cols.apply(lambda x: pd.Series(stats.zscore(x, nan_policy='omit')))
zscore_scipy = zscore_scipy.fillna(0)
zscore_scipy.columns = [col + '_zscore_scipy' for col in zscore_scipy.columns]

print("Z-scores using scipy.stats:")
print(zscore_scipy.head())
```

### Comparison of Methods

**Method 1 (Recommended)**: Clean, readable, and creates a separate function that can be reused.

**Method 2**: Most concise, good for one-time use.

**Method 3**: Most similar to your original loop approach, modifies DataFrame in place.

**Method 4**: Uses scipy.stats which is optimized and handles edge cases well.
