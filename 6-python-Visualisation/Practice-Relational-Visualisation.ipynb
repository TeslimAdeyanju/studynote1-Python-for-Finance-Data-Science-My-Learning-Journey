### Scatter Diagram 

```python
df.columns.to_list()
```

```python
sns.relplot(data = df, x='lifeExp', y = 'gdpPercap', kind='scatter')
plt.title("Life Exp vs GDP")
plt.show()
```

## Line Plot

```python
sns.relplot(data = df, x = 'year', y = 'pop', kind='line')
plt.title('Pop vs Year')
plt.show()
```
